```python
#functions to clean data

def convertValueToFloatOrNaN(stringVal):
    
    #if contains alphabets e.g "(D)" replace with NaN to be removed later
    if type(stringVal)==str and any(c.isalpha() for c in stringVal):
        return np.nan
    
    #to tackle values with commas e.g 5,000
    elif type(stringVal)==str and "," in stringVal:
        stringVal=stringVal.replace(",","")        
    return float(stringVal) #return float to account for values which may be in float

def removeDecimalYear(stringVal):
    stringVal=int(stringVal)
    return stringVal


def cleanData(df):
    
    valueName=df.columns[1] #to get the name of the column containing the variable (which is not year) 

    df[valueName]=df[valueName].apply(convertValueToFloatOrNaN)
    #print("-----------------convert variable to float or NaN dataframe-------------- \n",df)
    
    df=df.dropna() 
    
    
    no_duplicates=df.drop_duplicates(subset=['Year'],keep=False) #to find years that are not repeated
    #print("-----------------unduplicated years dataframe-------------- \n",no_duplicates)

    duplicated_years=set(df["Year"])-set(no_duplicates["Year"])#to find set of years that are repeated
    #print("duplicated yrs: ",duplicated_years)

    
    final_df=df.drop_duplicates(subset=['Year'],keep="first")#keep first occurence for ea yr & replace with avg values across row of same yr
    #print("-----------------dropped duplicates dataframe-------------- \n",final_df)

    for year in duplicated_years:
        duplicatedYr=df.loc[df['Year']==year] #get df of all rows with that specific yr to find avg of column
        #print(f"--------duplicated rows in year {year} ---------- \n", duplicatedYr)
        
        avg=duplicatedYr[valueName].mean()
        #print("avg: ",avg)
        
        final_df.loc[df['Year']==year,valueName]=avg
        #print(f"currently replacing {year}")
    #print("-----------------replacement with avg value for row of duplicated dataframe-------------- \n",final_df)

    final_df["Year"]=final_df["Year"].apply(removeDecimalYear) #fix year formatting
    #print("-----------------final dataframe-------------- \n",final_df)
    return final_df
```

```python
#function used in building linear regression model

def normalize_z(df):
    sd=df.std()
    mean=df.mean()
    df=(df-mean)/sd
    return df
    
    
#get the feature and target columns in format of dataframe
def get_features_targets(df, feature_names, target_names):
    df_feature=pd.DataFrame(df.loc[:,feature_names])
    df_target=pd.DataFrame(df.loc[:,target_names])
    
    return df_feature, df_target

#get feature columns in terms of numpy
def prepare_feature(df_feature):
    ###
    ### YOUR CODE HERE
    ###
    np_feature=df_feature.to_numpy()
    length=df_feature.shape[0]
    
   
    ones=np.ones((length,1),dtype=int)
    
    
    
    returnVal=np.concatenate((ones,np_feature ), axis=1)
    
    return returnVal

#get target column in terms of numpy
def prepare_target(df_target):
    ###
    ### YOUR CODE HERE
    ###
    np_target=df_target.to_numpy()
    
    return np_target

#return single column of predictedy values
def predict(df_feature, beta):
    ###
    ### YOUR CODE HERE
    ###
    df_feature=normalize_z(df_feature)
    np_feature=prepare_feature(df_feature)
    predictedy=calc_linear(np_feature, beta)
    return predictedy

def calc_linear(X, beta):
    ###
    ### YOUR CODE HERE
    ###
    return np.matmul(X,beta)

def split_data(df_feature, df_target, random_state=None, test_size=0.3):
    ###
    ### YOUR CODE HERE
    ###
    testing_size=int(df_feature.shape[0]*test_size)

    
    if random_state!=None:
        np.random.seed(random_state)
    testing_index=np.random.choice(df_feature.index,testing_size,replace=False)

    training_index=set(df_feature.index)-set(testing_index)


    df_feature_train=df_feature.loc[training_index]
    df_feature_test=df_feature.loc[testing_index]
    df_target_test=df_target.loc[testing_index]
    df_target_train=df_target.loc[training_index]

    return df_feature_train, df_feature_test, df_target_train, df_target_test,testing_index
  
def r2_score(y, ypred):
    errorSq=np.matmul((y-ypred).T,(y-ypred))
    SSres=np.sum(errorSq)
    ymean=np.mean(y)
    errorSq=np.matmul((y-ymean).T,(y-ymean))
    SStot=np.sum(errorSq)
    return 1-(SSres/SStot)


def compute_cost(X, y, beta):
   
    m=y.shape[0]
    error=calc_linear(X,beta)-y
    J=(1/(2*m))*np.matmul(error.T,error)
    return J
def gradient_descent(X, y, beta, alpha, num_iters):
    ###
    ### YOUR CODE HERE
    ###
    J_storage=[]
    length=X.shape[0]
    for i in range(num_iters):
        error=calc_linear(X, beta)-y

        beta-=alpha/length*np.matmul(X.T,error)
        J_storage.append(int(compute_cost(X,y,beta)))

    return beta, J_storage
```

```python
#putting all the above functions together


def buildModel(df,crop,featuresList,targetStr):
    
    numOfFeatures=len(featuresList)
    df_features, df_target = get_features_targets(df,featuresList,targetStr)
    df_features_train, df_features_test, df_target_train, df_target_test, testing_index = split_data(df_features, df_target,random_state=100,test_size=0.3)

# Normalize the features using z normalization
    df_features_train_z = normalize_z(df_features_train)

# Change the features and the target to numpy array using the prepare functions
    X = prepare_feature(df_features_train_z)
    target = prepare_target(df_target_train)

    iterations = 1500
    alpha = 0.01
    beta = np.zeros((numOfFeatures+1,1))

# Call the gradient_descent function
    beta, J_storage = gradient_descent(X,target,beta,alpha,iterations)

# call the predict() method
    pred = predict(df_features_test,beta)
    
 #data visualization:   
    print("final coefficients respective to order of input",featuresList," are: \n" ,beta, "and final cost is:", J_storage[-1])
    
    plt.scatter(df.loc[list(testing_index),"Year"],df_target_test)
    plt.scatter(df.loc[list(testing_index),"Year"],pred)
    
    
    
    plt.xlabel("Year" )
    plt.ylabel(f"Price of {crop} ")
    plt.legend(["actual price","predicted price"])
    plt.show()
    
```

```python
# put Python code to evaluate the model and to visualize its accuracy

def adjusted_r2_score(y,ypred,num_indp_var):  
    r2=r2_score(y, ypred)
    sample_size=y.size
    numerator=(1-r2)*(sample_size-1)
    denom=sample_size-num_indp_var-1
    return 1-(numerator/denom)
```
